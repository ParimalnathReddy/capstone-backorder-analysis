```python
def calculate_rolling_avg_backorder_rate(df):
    """Calculate the 6-month rolling average backorder rate for each part and warehouse efficiently."""
    # Ensure the DataFrame is sorted by 'issue_date'
    df = df.sort_values('issue_date')

    # Use a rolling window to calculate the average, leveraging the time-based window functionality
    df['avg_backorder_rate_6m'] = (
        df.set_index('issue_date')
        .groupby(['part', 'warehouse'])['back_order_flag']
        .rolling('180D')  # Time-based window of 6 months (180 days)
        .mean()
        .reset_index(drop=True)
    )

    return df

# Apply the function to the entire DataFrame
orders_2008 = calculate_rolling_avg_backorder_rate(orders_2008)
```

```python
def calculate_rolling_order_count(df):
    """Calculate the number of orders in the past 6 months for each part and warehouse efficiently."""
    # Ensure the DataFrame is sorted by 'issue_date'
    df = df.sort_values('issue_date')

    # Use a rolling window to count the number of orders in the past 6 months
    df['order_count_6m'] = (
        df.set_index('issue_date')
        .groupby(['part', 'warehouse'])['issue_quantity']
        .rolling('180D')  # Time-based window of 6 months (180 days)
        .count()
        .reset_index(drop=True)
    )

    return df

# Apply the function to your DataFrame
orders_2008 = calculate_rolling_order_count(orders_2008)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import statsmodels.api as sm

# Assuming 'orders_2008' is your DataFrame and 'churn_rate' is your target column
# Select all feature columns except the target
features = orders_2008.drop(columns=['back_order_flag']).columns

# DataFrame to store results
results = pd.DataFrame(columns=['Feature', 'Coefficient', 'Intercept', 'P-value'])

# Iterate over each feature and fit logistic regression
for feature in features:
    # Define the independent variable (feature) and the dependent variable (target)
    X = orders_2008[[feature]]
    y = orders_2008['back_order_flag']

    # Add constant term for intercept in statsmodels
    X_with_constant = sm.add_constant(X)

    # Fit logistic regression using statsmodels
    model = sm.Logit(y, X_with_constant).fit(disp=0)  # disp=0 suppresses output

    # Extract coefficient, intercept, and p-value
    coefficient = model.params[feature]
    intercept = model.params['const']
    p_value = model.pvalues[feature]

    # Append results to the DataFrame
    results = results.append({'Feature': feature, 'Coefficient': coefficient, 'Intercept': intercept, 'P-value': p_value}, ignore_index=True)

# Display results sorted by p-value (smallest p-value first)
results = results.sort_values(by='P-value').reset_index(drop=True)
print(results)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Select the columns for modeling
model_data = orders_2008[['asset_type', 'issue_priority', 'backorder_quantity', 'backorder_month',
                          'location', 'order_type', 'warehouse', 'avg_backorder_cost',
                           'part_backorder_rate', 'time_since_last_issue',
                          'warehouse_backorder_rate', 'avg_backorder_rate_6m', 'order_count_6m',
                          'back_order_flag']]

# 1. Encode categorical variables
categorical_cols = ['asset_type', 'issue_priority', 'location', 'order_type', 'warehouse']
label_encoders = {}
for col in categorical_cols:
    label_encoders[col] = LabelEncoder()
    model_data[col] = label_encoders[col].fit_transform(model_data[col])

# 2. Handle missing values - fill numeric columns with median
model_data.fillna(model_data.median(), inplace=True)

# 3. Separate features (X) and target (y)
X = model_data.drop(columns=['back_order_flag'])
y = model_data['back_order_flag']

# 4. Standardize numeric features (optional but beneficial for logistic regression)
scaler = StandardScaler()
numeric_cols = X.select_dtypes(include=['float64', 'int64']).columns
X[numeric_cols] = scaler.fit_transform(X[numeric_cols])
```

## Random Forest

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Train Logistic Regression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# Make predictions
y_pred_log = log_reg.predict(X_test)

# Evaluate the model
print("Logistic Regression Classification Report:")
print(classification_report(y_test, y_pred_log))
# Train Random Forest Classifier
rf_clf = RandomForestClassifier(random_state=42)
rf_clf.fit(X_train, y_train)

# Make predictions
y_pred_rf = rf_clf.predict(X_test)

# Evaluate the model
print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract feature importance from the Random Forest model
feature_importances = pd.DataFrame({'Feature': X.columns, 'Importance': rf_clf.feature_importances_})
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)

# Plot feature importances
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importances)
plt.title('Feature Importance from Random Forest')
plt.show()
```

## Data PreProssesing

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Select features for regression (excluding `backorder_duration` and `back_order_flag`)
regression_data = orders_2008[['asset_type', 'issue_priority', 'backorder_quantity', 'backorder_month',
                               'location', 'order_type', 'warehouse', 'avg_backorder_cost',
                               'part_backorder_rate', 'time_since_last_issue',
                               'warehouse_backorder_rate', 'avg_backorder_rate_6m', 'order_count_6m']][orders_2008['backorder_duration']>=0]

# Target variable
target = orders_2008['backorder_duration'][orders_2008['backorder_duration']>=0]

# 1. Encode categorical variables
categorical_cols = ['asset_type', 'issue_priority', 'location', 'order_type', 'warehouse']
label_encoders = {}
for col in categorical_cols:
    label_encoders[col] = LabelEncoder()
    regression_data[col] = label_encoders[col].fit_transform(regression_data[col])

# 2. Handle missing values by filling with median (if any)
regression_data.fillna(regression_data.median(), inplace=True)

# 3. Standardize numeric features (optional, beneficial for linear models)
scaler = StandardScaler()
numeric_cols = regression_data.select_dtypes(include=['float64', 'int64']).columns
regression_data[numeric_cols] = scaler.fit_transform(regression_data[numeric_cols])

# 4. Split data into features (X) and target (y)
X = regression_data
y = target
```

## Model building

###Linear Regression

```python
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Train Linear Regression model
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Make predictions
y_pred_lin = lin_reg.predict(X_test)

# Evaluate the model
print("Linear Regression Performance:")
print("MAE:", mean_absolute_error(y_test, y_pred_lin))
print("MSE:", mean_squared_error(y_test, y_pred_lin))
print("R:", r2_score(y_test, y_pred_lin))
```

###Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

# Train Random Forest Regressor model
rf_reg = RandomForestRegressor(random_state=42)
rf_reg.fit(X_train, y_train)

# Make predictions
y_pred_rf = rf_reg.predict(X_test)

# Evaluate the model
print("Random Forest Regressor Performance:")
print("MAE:", mean_absolute_error(y_test, y_pred_rf))
print("MSE:", mean_squared_error(y_test, y_pred_rf))
print("R:", r2_score(y_test, y_pred_rf))
```

### Can we have a single part with multiple priorities? For each part identify no.of back orders group by priority and also percentage of total group by priority.


```python
# prompt: part, count(*) group by priority,part in to column High, Medium, Low as per the value  and also percentage of total group by priority,part into columns High%, Medium%,Low%.

# Group by 'part' and 'issue_priority', then count the number of backorders
part_priority_counts = orders_2008.groupby(['part', 'issue_priority'])['backorder_quantity'].sum().reset_index()

# Pivot the table to create columns for High, Medium, Low
pivot_table = part_priority_counts.pivot_table(index='part', columns='issue_priority', values='backorder_quantity', aggfunc='sum', fill_value=0)

# Rename the columns
pivot_table = pivot_table.rename(columns={'High': 'High', 'Medium': 'Medium', 'Low': 'Low'})


# Calculate the total backorders for each part
pivot_table['Total'] = pivot_table.sum(axis=1)

# Calculate the percentage of total for each priority
for priority in ['High', 'Medium', 'Low']:
  if priority in pivot_table.columns:
    pivot_table[priority + '%'] = (pivot_table[priority] / pivot_table['Total']) * 100

# Remove Total Column
pivot_table = pivot_table[['High', 'Medium', 'Low', 'High%', 'Medium%', 'Low%']]

# Fill NaN with 0
pivot_table = pivot_table.fillna(0)

pivot_table
```
